```python
import numpy as np # importando para manipular dados (ex: usaremos para atribuir valores nulos)

# Criando df de teste
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# fillna -> preencher valores nulos
# round2 -> arredondando para duas casas decimais
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))  # mdia
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())  # mediana
df_salarios
```

```python
# Criando outro df para testes
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# Fazer preenchimento, criando nova coluna
# ffill(foward fill) -> repetiu o anterior para os nulos
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas

# Se segunda fosse NaN, daria problema
```

```python
# Criando outro df para testes
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [np.nan, np.nan, np.nan, 28, 27]
})

# Fazer preenchimento, criando nova coluna
# bfill(back fill) -> repetiu o prximo para os nulos
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
# Fazer teste com cidade
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['So Paulo', np.nan, 'Curitiba', np.nan, 'Belm']
})

# Colocando no informado para os vazios
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna("No Informado")
df_cidades
```

## Grfico de Barras Interativo

```python
# Biblioteca para deixar os grficos mais interativos
import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('senoridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senoridade',
             y='usd',
             title='Mdia Salarial por Senioridade',
             labels={'senoridade': 'Nvel de Senioridade', 'usd': 'Mdia Salarial Anual (USD)'})
fig.show()
```

```python
# Exibir label na porcentagem
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporo dos tipos de trabalho',
             hole=0.5

)

# Adicionando aqui
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
import pycountry

# Funo para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

# Criar nova coluna com cdigo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular mdia salario por pas (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salrio mdio de Cientista de Dados por pas',
                    labels={'usd': 'Salrio mdio (USD)', 'residencia_iso3': 'Pais'})

fig.show()
```
